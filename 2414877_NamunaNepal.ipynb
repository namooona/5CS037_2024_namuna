Task - 3 - Outlier Detection:
Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5 × IQR rule.

```python
import pandas as pd

#Identify outlier countries in South Asia based on their Score and GDP per Capita
# Calculating IQR for score column of south asia
score_q1= south_asia["score"].quantile(0.25)
score_q3= south_asia["score"].quantile(0.75)
score_iqr= score_q3-score_q1
print("IQR for score is:")
print(score_iqr)

# Calculating IQR for GDP column of south asia
gdp_q1= south_asia["Log GDP per capita"].quantile(0.25)
gdp_q3= south_asia["Log GDP per capita"].quantile(0.75)
gdp_iqr= gdp_q3-gdp_q1
print("IQR for GDP is:")
print(gdp_iqr)

print()
#Define outliers using the 1.5 × IQR rule
print("For score column:")
score_lower_bound= score_q1-1.5*score_iqr
score_upper_bound= score_q3+1.5*score_iqr
print("Lower bound for score is:")
print(score_lower_bound)
print("Upper bound for score is:")
print(score_upper_bound)

print()

print("For GDP column:")
gdp_lower_bound= gdp_q1-1.5*gdp_iqr
gdp_upper_bound= gdp_q3+1.5*gdp_iqr
print("Lower bound for GDP is:")
print(gdp_lower_bound)
print("Upper bound for GDP is:")
print(gdp_upper_bound)

#Identifying outliers for score
outliers_score= south_asia[(south_asia["score"]<score_lower_bound) | (south_asia["score"]>score_upper_bound)]
print("Outliers for score are:")
print(outliers_score)

print()
#Identifying outliers for GDP
outlier_gdp= south_asia[(south_asia["Log GDP per capita"]<gdp_lower_bound) | (south_asia["Log GDP per capita"]>gdp_upper_bound)]
print("Outliers for GDP are:")
print(outlier_gdp)

#Combining both score and gdp outliers
outlier= pd.concat([outliers_score, outlier_gdp]).drop_duplicates()
print("Combined outliers are:")
print(outlier)
```

6. Outlier Detection:
• Identify outlier countries in both regions based on Score and GDP per Capita.

```python
import pandas as pd

# Identify outliers of South Asia using score and GDP per capita
# Calculate IQR for Score in South Asia
score_q1_south_asia = south_asia["score"].quantile(0.25)
score_q3_south_asia = south_asia["score"].quantile(0.75)
score_iqr_south_asia = score_q3_south_asia - score_q1_south_asia
score_lower_bound_south_asia = score_q1_south_asia - 1.5 * score_iqr_south_asia
score_upper_bound_south_asia = score_q3_south_asia + 1.5 * score_iqr_south_asia
print("The lower bound for score in South Asia is:")
print(score_lower_bound_south_asia)
print("The upper bound for score in South Asia is:")
print(score_upper_bound_south_asia)
print()

# Calculate IQR for GDP per Capita in South Asia
gdp_q1_south_asia = south_asia["Log GDP per capita"].quantile(0.25)
gdp_q3_south_asia = south_asia["Log GDP per capita"].quantile(0.75)
gdp_iqr_south_asia = gdp_q3_south_asia - gdp_q1_south_asia
gdp_lower_bound_south_asia = gdp_q1_south_asia - 1.5 * gdp_iqr_south_asia
gdp_upper_bound_south_asia = gdp_q3_south_asia + 1.5 * gdp_iqr_south_asia
print("The lower bound for GDP per capita in South Asia is:")
print(gdp_lower_bound_south_asia)
print("The upper bound for GDP per capita in South Asia is:")
print(gdp_upper_bound_south_asia)
print()

# Identify outliers for Score and GDP per Capita in South Asia
south_asia_score_outliers = south_asia[(south_asia["score"] < score_lower_bound_south_asia) | (south_asia["score"] > score_upper_bound_south_asia)]
south_asia_gdp_outliers = south_asia[(south_asia["Log GDP per capita"] < gdp_lower_bound_south_asia) | (south_asia["Log GDP per capita"] > gdp_upper_bound_south_asia)]
print("The South Asia score outliers are:")
print(south_asia_score_outliers)
print("The South Asia GDP per capita outliers are:")
print(south_asia_gdp_outliers)
print()

# Calculate IQR for Score in Middle East
score_q1_middle_east = middle_east["score"].quantile(0.25)
score_q3_middle_east = middle_east["score"].quantile(0.75)
score_iqr_middle_east = score_q3_middle_east - score_q1_middle_east
score_lower_bound_middle_east = score_q1_middle_east - 1.5 * score_iqr_middle_east
score_upper_bound_middle_east = score_q3_middle_east + 1.5 * score_iqr_middle_east
print("The lower bound for score in Middle East is:")
print(score_lower_bound_middle_east)
print("The upper bound for score in Middle East is:")
print(score_upper_bound_middle_east)
print()

# Calculate IQR for GDP per Capita in Middle East
gdp_q1_middle_east = middle_east["Log GDP per capita"].quantile(0.25)
gdp_q3_middle_east = middle_east["Log GDP per capita"].quantile(0.75)
gdp_iqr_middle_east = gdp_q3_middle_east - gdp_q1_middle_east
gdp_lower_bound_middle_east = gdp_q1_middle_east - 1.5 * gdp_iqr_middle_east
gdp_upper_bound_middle_east = gdp_q3_middle_east + 1.5 * gdp_iqr_middle_east
print("The lower bound for GDP per capita in Middle East is:")
print(gdp_lower_bound_middle_east)
print("The upper bound for GDP per capita in Middle East is:")
print(gdp_upper_bound_middle_east)
print()

# Identify outliers for Score and GDP per Capita in Middle East
middle_east_score_outliers = middle_east[(middle_east["score"] < score_lower_bound_middle_east) | (middle_east["score"] > score_upper_bound_middle_east)]
middle_east_gdp_outliers = middle_east[(middle_east["Log GDP per capita"] < gdp_lower_bound_middle_east) | (middle_east["Log GDP per capita"] > gdp_upper_bound_middle_east)]
print("The Middle East score outliers are:")
print(middle_east_score_outliers)
print("The Middle East GDP per capita outliers are:")
print(middle_east_gdp_outliers)
print()
```

• Plot these outliers and discuss their implications.

```python
# Scatter plot for South Asia
plt.figure(figsize=(10, 5))

# South Asia Scatter Plot: Score vs GDP per Capita
plt.subplot(1, 2, 1)
plt.scatter(south_asia["Log GDP per capita"], south_asia["score"], color="blue", alpha=0.7, label="Data")
# Highlight outliers for South Asia
plt.scatter(south_asia_score_outliers["Log GDP per capita"], south_asia_score_outliers["score"], color="red", label="Score Outliers", marker="o")
plt.scatter(south_asia_gdp_outliers["Log GDP per capita"], south_asia_gdp_outliers["score"], color="green", label="GDP Outliers", marker="x")
plt.xlabel('Log GDP per Capita', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.title('South Asia: Score vs GDP per Capita', fontsize=14)
plt.grid(True)
plt.legend()

# Scatter plot for Middle East
plt.subplot(1, 2, 2)
plt.scatter(middle_east["Log GDP per capita"], middle_east["score"], color="green", alpha=0.7, label="Data")
# Highlight outliers for Middle East
plt.scatter(middle_east_score_outliers["Log GDP per capita"], middle_east_score_outliers["score"], color="red", label="Score Outliers", marker="o")
plt.scatter(middle_east_gdp_outliers["Log GDP per capita"], middle_east_gdp_outliers["score"], color="blue", label="GDP Outliers", marker="x")
plt.xlabel('Log GDP per Capita', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.title('Middle East: Score vs GDP per Capita', fontsize=14)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
